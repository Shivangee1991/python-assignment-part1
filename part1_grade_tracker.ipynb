```python
#________Task 1: Data Parsing and Profile Cleaning for Student Records________
# Given Dataset
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

# Loop through all students for cleaning data like name, roll and marks and store in new list of dictionaries
cleaned_students = []
for student in raw_students:
    # Clean name by stripping whitespace and converting to title case
    cleaned_name = student["name"].strip().title()
    
    # Clean roll number by stripping whitespace
    cleaned_roll = student["roll"].strip()
    
    # Clean marks by splitting the string, stripping whitespace, and converting to integers
    cleaned_marks = [int(mark.strip()) for mark in student["marks_str"].split(",")]

    # Validate cleaned data
    is_valid = all(word.isalpha()for word in cleaned_name.split())
    
    # Create a new dictionary for the cleaned student data
    cleaned_student = {
        "name": cleaned_name,
        "roll": cleaned_roll,
        "marks": cleaned_marks
    }
    
    # Append the cleaned student dictionary to the list
    cleaned_students.append(cleaned_student)
    
    # print validation for cleaned data
    if is_valid:
        print(f"{cleaned_name}   → ✓ Valid name")
    else:
        print(f"{cleaned_name}   → ✗ Invalid name")
    
    #Print profile card for each student
    print("\n===================================================")
    print(f"Name: {cleaned_name}")
    print(f"Roll Number: {cleaned_roll}")
    print(f"Marks: {cleaned_marks}")
    print("===================================================\n")

    # find the student with roll number 103 and print name in all CAPS and lowercase
    for student in cleaned_students:
        if student["roll"] == "103":
            print("Uppercase Name for Roll Number 103: ", student["name"].upper())
            print("Lowercase Name for Roll Number 103: ", student["name"].lower())
            print("\n")
```

```python
#_______Task2: Mark Analysis usig loops and conditionals________

# Given Dataset for Mark Analysis

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
print(f"Mark Analysis for {student_name}")

# using for loop,print each subject alongside its marks and grade as per given scheme
def calculate_grade(mark):
    if 90 <= mark <= 100:
        grade = "A+"
    elif 80 <= mark < 90:
        grade = "A"
    elif 70 <= mark < 80:
        grade = "B"
    elif 60 <= mark < 70:
        grade = "C"
    else:
        grade = "F"
    return grade

# Loop through subjects and marks to print subject, mark and grade
print(f"{'Subject':<15}{'Marks':<8}{'Grade'}")
for i in range(len(subjects)):
    grade = calculate_grade(marks[i])
    print(f"{subjects[i]:<15}{marks[i]:<8}{grade}")

Total_marks = sum(marks)
Average_marks = Total_marks / len(marks)
Max_mark = max(marks)
Min_mark = min(marks)
max_mark_index = marks.index(Max_mark)
min_mark_index = marks.index(Min_mark)

print("\n  Summary:")
print("************************************************************")
print(f"Total Marks: {Total_marks}")
print(f"Average Marks: {Average_marks:.2f}")
print(f"Highest Mark: {Max_mark} in {subjects[max_mark_index]}")
print(f"Lowest Mark: {Min_mark} in {subjects[min_mark_index]}")
print("************************************************************")

# Use While loop for simulate a mark entry system that allows adding new subjecsts and marks until user decides to stop

print("\nMark Entry System")

new_subjects = []
while True:
    subject_input = input("Enter subject name: ").strip()
    if subject_input.lower() == "stop":
        break
    marks_input = input("Enter marks for the subject: ").strip()
    if not marks_input.isdigit():
        print("Invalid marks input. Please enter a number.\n")
        continue
# Add valid subject and marks to the list
# Convert marks input to integer and append to the list        
    subjects.append(subject_input)
    marks.append(int(marks_input))


    print(f"Added {subject_input} with marks {marks_input}.\n")
    print("After adding new subject and marks, Type 'stop' to end the mark entry system.\n")
    
```

```python
#__________Task3: Class Performance Summary________
# Given Dataset for Class Performance
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
# print statements for class performance summary
print("Name                | Average        | Status")
print("--------------------------------------------------")
pass_count = 0
fail_count = 0

Class_Topper_name = ""
Class_Topper_avg = 0

Sumof_Averages = 0

#For loop for name and avg marks of each student and pass/fail status
for name, marks in class_data:
    average = sum(marks) / len(marks)
    Sumof_Averages += average

# if-else for pass/fail status    
    if average >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

# if statement to find class topper
    if average > Class_Topper_avg:
          Class_Topper_avg = average
          Class_Topper_name = name
# print name, average and status in formatted way
    print(f"{name:<20} | {average:<14.2f} | {status}")    

# Calculate class average        
Class_Average = Sumof_Averages / len(class_data)

#Final summary print statements for class performance
print("\nClass Performance Summary:")
print("**************************************************")
print(f"Total Students: {len(class_data)}")
print(f"Passed: {pass_count}")
print(f"Failed: {fail_count}")
print(f"Class Topper: {Class_Topper_name} with an average of {Class_Topper_avg:.2f}")
print(f"Class Average: {Class_Average:.2f}")
```

```python
#__________Task4: String Manipulation Utility________
# Given String for manipulation
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# Strip leading and trailing whitespace from eassay
cleaned_essay = essay.strip()
print("1.> Cleaned Essay:\n", cleaned_essay)
print("\n")

# Convert essay to title case
title_case_essay = cleaned_essay.title()
print("2.> Essay in Title Case:\n", title_case_essay)
print("\n")

# Count Python words in the essay
python_count = cleaned_essay.lower().count("python")
print(f"3.> Count of 'Python' in the essay: {python_count}")
print("\n")

#Replace every occurrence of "python" in clean_essay with "Python 🐍
replaced_essay = cleaned_essay.replace("python", "Python 🐍")
print("4.> Essay after replacing 'python' with 'Python 🐍:\n",replaced_essay)
print("\n")

# Split the essay into sentences and print each sentence on a new line
sentences = [sentence.strip() for sentence in cleaned_essay.split(".") if sentence.strip()]
print("5.> Essay split into sentences:")
for sentence in sentences:
    print(sentence)

# print each statement in new line starting with 1. and end with full stop    
print("\n6.> Essay with each sentence numbered:")
for i, sentence in enumerate(sentences, start=1):
    ending = '' if sentence.endswith(('.', '?', '!')) else '.'
    print(f"{i}. {sentence}{ending}")
   
    
```
